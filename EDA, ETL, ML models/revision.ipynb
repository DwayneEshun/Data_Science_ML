```python
import pandas as pd
import numpy as np

census = pd.read_csv("census.csv")
census.columns
census.groupby('STNAME')['CENSUS2010POP'].agg(['max', 'min', 'mean']).round(2).sort_values(by = 'max', ascending = False)
```

```python
census.groupby('STNAME')['DEATHS2015'].agg(['min','max']).sort_values(by = 'min').head(10)
```

```python
#fetch data for all cities with 0 deaths

states = ['Hawaii', 'Montana', 'Nebraska', 'North Dakota', 'Texas']
census[census.STNAME.isin(states)].groupby('CTYNAME')['DEATHS2015'].agg(['min', 'max']).sort_values('min', ascending = True).head(10)
```

```python
census[census.STNAME.isin(states)].groupby(['STNAME', 'CTYNAME'])['DEATHS2015'].agg(['min', 'max']).sort_values(by = 'min').head(20)
```

```python
print('Maximum Births:',census.BIRTHS2015.max())
print('Minimum Births:', census.BIRTHS2015.min())
```
