```python
cars = pd.read_csv('Toyota.csv', index_col= 0, na_values= ["??", "????"])
cars.head(20)
```

```python
cars.info()
```

```python
cars.isnull().sum()
```

```python
cars.replace('three',3,inplace = True)
cars.replace('four',4,inplace = True)
cars.replace('five',5,inplace = True)
cars.head(20)
```

```python
cars.dtypes
```

```python
cars['Doors'] = cars['Doors'].astype('int64')
cars['MetColor'] = cars['MetColor'].astype('object')
cars['Automatic'] = cars['Automatic'].astype('object')
```

```python
cars.info()
```

```python
cars['Automatic'] = cars['Automatic'].astype('object')
cars['MetColor'] = cars['MetColor'].astype('object')
cars.dtypes
```

```python
cars.describe()
```

```python
cars['Age'].mean()
```

```python
cars['KM'].median()
```

```python
cars['HP'].mean()
```

```python
cars['FuelType'].mode()[0]
```

```python
cars['MetColor'].value_counts()
```

```python
cars['MetColor'].mode()[0]
```

```python
cars['Age'].fillna(cars['Age'].mean(), inplace = True)
cars['KM'].replace(np.nan,cars['KM'].median(), inplace = True)
cars['HP'].fillna(cars['HP'].mean(), inplace = True)
cars['MetColor'].fillna(cars['MetColor'].mode()[0], inplace = True)
cars['FuelType'].fillna(cars['FuelType'].mode()[0], inplace = True)
```
